```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt 
```

```python
df=pd.read_csv("/Data/covid_19_india.csv", parse_dates=["Date"], dayfirst=True)   
df.head() 
```

```python
df=df[["Date","State/UnionTerritory","Cured","Deaths","Confirmed"]]
df.head(3)  
```

```python
df.columns=["date","state","cured","deaths","confirmed"] 
df.head(3) 
```

```python
df.sort_values(by='date',ascending=True) 
```

```python
# How confirmed cases increases with time 

sns.lineplot(x="date", y="confirmed",data=df, color='g') 
plt.show()
```

```python
today = df[df.date=="2021-08-11"]
```
